### Year Built

```python
#function to categorize year built
def categorize_year_decade(year):
    if year < 1900 :
        return '1800s'
    elif year < 1910 and year> 1900:
        return '1900s'
    elif year < 1920and year> 1910:
        return '1910s'
    elif year < 1930 and year> 1920:
        return '1920s'
    elif year < 1940 and year> 1930:
        return '1930s'
    elif year < 1950 and year> 1940:
        return '1940s'
    elif year < 1960 and year> 1950:
        return '1950s'
    elif year < 1970 and year> 1960:
        return '1960s'
    elif year < 1980 and year> 1970:
        return '1970s'
    elif year < 1990 and year> 1980:
        return '1980s'
    elif year < 2000 and year> 1990:
        return '1990s'
    else:
        return '2000s'

#function to create a new column 'year_decade'
df['year_decade'] = df['yr_built'].apply(categorize_year_decade)

# Group by 'year_decade'
df.groupby('year_decade')[['price', 'sqft_living', 'sqft_lot', 'grade', 'floors']].mean()
```

- The analysis provides insights into how the decade of year built is associated with various attributes like price, living area size, lot size, grade, and floors.
- Generally, newer houses tend to have larger living areas, higher lot sizes, higher grades, and more floors.
- Houses from the 1990s and 2000s have higher average grades and living area sizes compared to earlier decades, which could reflect advancements in design and construction quality.
- While there are fluctuations, overall, the prices have increased over the years, with more recent decades showing higher mean prices.

## Price_per_sqft

```python
#price_per_sqft function
def price_per_sqft(row):
    return row['price'] / row['sqft_living']

# Apply the price_per_sqft function
df['price_per_sqft'] = df.apply(price_per_sqft, axis=1)
```

```python
df.head()
```

```python
#median price per square foot
median = df['price_per_sqft'].median()

#mean price per square foot for different renovation
renovated = df[df['renovated'] == 'Renovated']['price_per_sqft'].mean()
not_renovated = df[df['renovated'] == 'Not Renovated']['price_per_sqft'].mean()
```

```python
print("Median Price Per Square Foot:", median)
print("Mean Price Per Square Foot (Renovated):", renovated)
print("Mean Price Per Square Foot (Not Renovated):", not_renovated)
```

- Renovated houses, on average, have a higher price per square foot than non-renovated houses. This supports the assumption that renovations contribute to increased property values.

```python
#mean price per square foot for different year decades
df.groupby('year_decade')['price_per_sqft'].mean()
```

- **Early 20th century:** Homes built in the early 20th century (1900s and 1910s) had a higher mean price per square foot than homes built in other decades. This suggests that there has been a historical trend of premium pricing for older homes.
- **Mid-20th century:** The mean price per square foot saw a decline during the mid-20th century (1940s and 1950s). This may have been due to changes in housing preferences or market dynamics during that period.
- **Recent years:** In more recent years (2000s), the mean price per square foot has shown an uptick. This suggests that there is renewed interest and increased value for properties built in the 21st century.
